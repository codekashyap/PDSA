## Timing our code

#### How long does it take for our code to execute?
 - Depends on the language

#### Python has a library `time` with various useful functions

#### `perf_time()` is a performance counter
 - Absolute value of `perf_time()` is not meaningful
 - Compare 2 consecutive readings to get an interval
 - Default unit is seconds
 - Python usually executes 10^7 operations per second
 

```python
import time

start = time.perf_counter()
...
# Execute some code
...
end = time.perf_counter()

elapsed = end - start
```

### A Timer object

```python
import time

class Timer:
  def __init__(self):
    self._start = 0
    self._elapsed = 0
  
  def start(self):
    self._start = time.perf_counter()
  
  def stop(self):
    self._elapsed = time.perf_counter() - self._start
  
  def elapsed(self):
    return self._elapsed
```
