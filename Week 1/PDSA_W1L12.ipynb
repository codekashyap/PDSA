### GCD using naive algorithm

```python
def gcd(m, n):
  common_factors = []

  for i in range(1, min(m, n) + 1):
    if m % i == 0 and n % i == 0:
      common_factors.append(i)
  
  return common_factors[-1]
```

```python
gcd(567812346, 876543217)
```

### Recursive GCD using m - n

```python
def gcd(m, n):
  a, b = max(m, n), min(m, n)

  if a % b == 0:
    return b
  
  return gcd(b, a - b)
```

```python
gcd(2, 99999999)
```

### Euclid's algorithm

```python
def gcd(m, n):
  a, b = max(m, n), min(m, n)

  if a % b == 0:
    return b
  
  return gcd(b, a % b)
```
