## Using Exceptions 'positively'

### Traditional approach

```python
if b in scores.keys():
  scores[b].append(s)
else:
  scores[b] = [s]
```

### Using exceptions

```python
try:
  scores[b].append(s)
except KeyError:
  scores[b] = [s]
```
